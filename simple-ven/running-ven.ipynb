# Home Behavior

Homes have random load shapes and random requests for increased load capacity of 6kW

```python
# simulating 10 homes
ven = HomeLoad(intervals = 24)
homes = [HomeLoad(intervals = 24) for i in range(0,10)]
    
```

```python
plot_home(homes[0])
```

```python
plot_many_homes(homes)
```

```python
# dummy responses from VTN 
for home in homes:
    # clearing the reservation load shape
    home.clear_reservation_load_shape()
    # vtn response is simple list of prices, that is the same length as the capacity request
        # alternative could be list of tuples with time and price(like the capacity reques itself)
    # print('HOME CAPACITY REQUEST (times, kW)', home.capacity_request)
    vtn_response = [random.choice([0,0.05, 0.1, 0.15,0.2]) for i in home.capacity_request]
    # print('VTN CAPACITY PRICES ($/kW)', vtn_response)
    home.adjust_capacity_request(vtn_response)
```
